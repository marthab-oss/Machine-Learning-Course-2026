Dataset: Sequential Mediation from Kaggle



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
print(os.getcwd())
```

```python
#import datafile df with pandas
df=pd.read_csv('sequential_mediation_dataset.csv')
df
```

```python
#overview of data
#sort all coloumns
cols=['educational_intervention','coping_strategy_improvement', 'social_support_enhancement', 'mental_health_score', 'socioeconomic_status','school_quality_score', 'baseline_cognitive_score'] #, 'random_noise'
df_0=df[df['educational_intervention']==0]
df_1=df[df['educational_intervention']==1]
for a in cols:
  sorted_0=df_0[a].sort_values().reset_index(drop=True)
  sorted_1=df_1[a].sort_values().reset_index(drop=True)
  #distribution of the 5 characteristics (w/o random noise) to intervention 0 or 1
  plt.plot(sorted_0, label=f'{a}(0)')
  plt.plot(sorted_1, label=f'{a}(1)')



#df_0[cols].plot(kind='box', figsize=(15,4))
#df_1[cols].plot(kind='box', figsize=(15,4))

plt.tight_layout()

plt.show()
```
